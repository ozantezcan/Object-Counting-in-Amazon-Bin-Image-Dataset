```python
import pandas
from matplotlib import pyplot as plt
import numpy as np
% matplotlib inline

acc = pandas.read_csv('./run_resnet18__sgd_0.0xsingle + 1.0Xmulti_June13  16_50_51-tag-train_accuracy.csv')
eps = acc.values[:,-2]
tr_acc = acc.values[:,-1]
acc = pandas.read_csv('./run_resnet18__sgd_0.0xsingle + 1.0Xmulti_June13  16_50_51-tag-val_accuracy.csv')
val_acc = acc.values[:,-1]

ccr1 = pandas.read_csv('./run_resnet18__sgd_0.0xsingle + 1.0Xmulti_June13  16_50_51-tag-train_CIR-1.csv')
tr_ccr1 = ccr1.values[:,-1]
ccr1 = pandas.read_csv('./run_resnet18__sgd_0.0xsingle + 1.0Xmulti_June13  16_50_51-tag-val_CIR-1.csv')
val_ccr1 = ccr1.values[:,-1]


# Create plots with pre-defined labels.
# Alternatively, you can pass labels explicitly when calling `legend`.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(20,10))
ax1.plot(eps, tr_acc, 'r', label='train')
ax1.plot(eps, val_acc, 'k', label='validation')
# Now add the legend with some customizations.
legend = ax1.legend()

# The frame is matplotlib.patches.Rectangle instance surrounding the legend.
frame = legend.get_frame()
frame.set_facecolor('0.90')

# Set the fontsize
for label in legend.get_texts():
    label.set_fontsize(20)

for label in legend.get_lines():
    label.set_linewidth(1.5)  # the legend line width

ax2.set_xlabel('epoch', fontsize = 16)
ax2.set_ylabel('$CCR_1$', fontsize = 16)

ax2.plot(eps, tr_ccr1, 'r', label='train')
ax2.plot(eps, val_ccr1, 'k', label='validation')
# Now add the legend with some customizations.
legend = ax2.legend()

# The frame is matplotlib.patches.Rectangle instance surrounding the legend.
frame = legend.get_frame()
frame.set_facecolor('0.90')

# Set the fontsize
for label in legend.get_texts():
    label.set_fontsize(20)

for label in legend.get_lines():
    label.set_linewidth(1.5)  # the legend line width

ax1.set_xlabel('epoch', fontsize = 16)
ax1.set_ylabel('CCR', fontsize = 16)
```
